```python
import pandas as pd
import os
import datetime
from tqdm.notebook import tqdm
import numpy as np
```

## EXTRACT
Store datasets into DataFrames

```python
csv_data = "../webscraping/sf_cl_webscrape.csv"
sf_cl_df = pd.read_csv(csv_data)
```

```python
sf_cl_df
```

## TRANSFORM
Clean DataFrame

```python
sf_df = sf_cl_df.drop(columns=["Unnamed: 0"])
sf_df.head()
```

```python
sf_df = sf_df.rename(columns = {"datetimes": "datetime", "hoods": "neighborhood", "prices": "rent",\
                                                "bedrooms": "br", "bathrooms": "ba", "housing_type": "housingType",\
                                                "titles": "title", "descriptions": "description", "links": "link"})
sf_df.head()
```

```python
# drop duplicate rows
sf_df = sf_df.drop_duplicates(subset='link')
len(sf_df.drop_duplicates(subset='link'))
```

```python
# drop duplicate rows
sf_df = sf_df.drop_duplicates(subset='title')
len(sf_df.drop_duplicates(subset='title'))
```

```python
sf_df.info()
```

```python
# convert string to datetime

sf_df['datetime'] = pd.to_datetime(sf_df['datetime'])
sf_df['datetime']
```

```python
sf_df['rent'] = sf_df['rent'].str.replace(',', '')
sf_df['rent'] = sf_df['rent'].str.replace('$', '')
sf_df['rent'] = pd.to_numeric(sf_df['rent'])
sf_df['rent']
```

```python
# set min and max for rent and drop rows

sf_df = sf_df.loc[(sf_df['rent'] > 500.00) & (sf_df['rent'] < 7999.00)]
```

```python
pd.set_option('display.max_rows', None)
sf_df
```

```python
sf_df.describe()
```

```python
sf_df['br'] = sf_df['br'].str.replace('BR', '')
sf_df['br'] = pd.to_numeric(sf_df['br'])
```

```python
sf_df['ba'].value_counts()
```

```python
sf_df['ba'] = sf_df['ba'].str.replace('splitBa', '1')
sf_df['ba'] = sf_df['ba'].str.replace('sharedBa', '1')
sf_df['ba'] = sf_df['ba'].str.replace('Ba', '')
sf_df['ba'] = pd.to_numeric(sf_df['ba'])
```

```python
sf_df['sqft'] = sf_df['sqft'].str.replace('none', '')
sf_df['sqft'] = pd.to_numeric(sf_df['sqft'])
```

```python
sf_df['cats'] = sf_df['cats'].str.replace('none', 'no')
```

```python
sf_df['dogs'] = sf_df['dogs'].str.replace('none', 'no')
sf_df['dogs'].value_counts()
```

```python
sf_df['furnished'] = sf_df['furnished'].str.replace('none', 'no')
sf_df['furnished'].value_counts()
```

```python
# drop duplicate rows after cleaning up titles
sf_df = sf_df.drop_duplicates(subset='title')
len(sf_df.drop_duplicates(subset='title'))
```

```python
sf_df.loc[:, 'description'] = sf_df['description'].str.replace('\n\nQR Code Link to This Post\n\n\n', '').copy()
sf_df.loc[:, 'description'] = sf_df['description'].str.replace('\n', '').copy()
sf_df['description']
```

```python
sf_df.info()
```

```python
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.title().copy()
sf_df['neighborhood'].value_counts()
```

```python
# data cleaning neighborhood column

sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('(', '')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace(')', '')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.lstrip()

sf_df.loc[sf_df['neighborhood'].str.contains('Central Sunset'), 'neighborhood'] = 'Sunset / Parkside'
sf_df.loc[sf_df['neighborhood'].str.contains('Inner Sunset'), 'neighborhood'] = 'Inner Sunset / Ucsf'
sf_df.loc[sf_df['neighborhood'].str.contains('Outer Sunset'), 'neighborhood'] = 'Sunset / Parkside'
sf_df.loc[sf_df['neighborhood'].str.contains('Soma/South Beach'), 'neighborhood'] = 'Soma / South Beach'
sf_df.loc[sf_df['neighborhood'].str.contains('Marina/Cow Hollow'), 'neighborhood'] = 'Marina / Cow Hollow'

sf_df.loc[sf_df['neighborhood'].str.contains('Pacific Height'), 'neighborhood'] = 'Lower Pac Hts'
sf_df.loc[sf_df['neighborhood'].str.contains('Lower Pacific Heights'), 'neighborhood'] = 'Lower Pac Hts'

sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('1532 Ne 21St Avenue', 'Inner Sunset / Ucsf')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('San Francisco Sunset', 'Sunset / Parkside')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('Civic Center', 'Downtown / Civic / Van Ness')

sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('Usf', 'USF')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('Sfsu', 'SFSU')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('Ccsf', 'CCSF')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('Soma', 'SOMA')
sf_df.loc[:, 'neighborhood'] = sf_df['neighborhood'].str.replace('Ucsf', 'UCSF')


sf_df['neighborhood'].value_counts()
```

```python
# drop rows where neighborhood is outside of sf area

non_sf = ["Santa Cruz", "Sunnyvale", "Belmont", "Southern Hills", "San Ramon",\
            "Redwood City", "Oakland", "Suisun City", "Santa Rosa", "Pacifica", "Berkeley", "13040 Brookpark Rd, Oakland, Ca",\
            "Crocker Amazon", "Colusa Ct,San Bruno, Ca"]

for i in tqdm(non_sf): 
    filt = sf_df[sf_df['neighborhood'].str.contains(i)]
    sf_df.drop(index = filt.index, inplace = True)
```
